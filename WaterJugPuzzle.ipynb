```python
class WaterJugPuzzle:
    def __init__(self):
        self.jugFive = 0
        self.jugThree = 0
#-------------------------------------------------------------------------------
# Fill the jugs
#-------------------------------------------------------------------------------
    def fill_five(self):
        self.jugFive = 5
        print("Fill jugFive to 5")

    def fill_three(self):
        self.jugThree = 3
        print("Fill jugThree to 3")
#-------------------------------------------------------------------------------
# Empty the jugs
#-------------------------------------------------------------------------------
    def empty_five(self):
        self.jugFive = 0
        print("Empty jugFive")

    def empty_three(self):
        self.jugThree = 0
        print("Empty jugThree")
#-------------------------------------------------------------------------------
# Pour the contants of the jug into the other
#-------------------------------------------------------------------------------
    def pour_three_to_five(self):
        space = 5 - self.jugFive
        transfer = min(space, self.jugThree)
        self.jugFive += transfer
        self.jugThree -= transfer
        print(f"Pour {transfer} from jugThree to jugFive")

    def pour_five_to_three(self):
        space = 3 - self.jugThree
        transfer = min(space, self.jugFive)
        self.jugFive -= transfer
        self.jugThree += transfer
        print(f"Pour {transfer} from jugFive to jugThree")
#-------------------------------------------------------------------------------
# Check for solved state
#-------------------------------------------------------------------------------
    def is_solved(self):
        return self.jugFive == 4

    def print_state(self):
        print(f"jugFive: {self.jugFive}L, jugThree: {self.jugThree}L")
#-------------------------------------------------------------------------------
# Use the known steps to solve the puzzle
#-------------------------------------------------------------------------------
    def solve(self):
        self.fill_five()
        self.pour_five_to_three()
        self.empty_three()
        self.pour_five_to_three()
        self.fill_five()
        self.pour_five_to_three()

        print("Final State:")
        self.print_state()
#-------------------------------------------------------------------------------
# Run the puzzle solver
#-------------------------------------------------------------------------------
wjp = WaterJugPuzzle()
wjp.solve()
```
